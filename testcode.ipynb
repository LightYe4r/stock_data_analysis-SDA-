```python
from bs4 import BeautifulSoup
import requests

# 크롤링할 웹 페이지 URL
url = 'http://comp.fnguide.com/'

# 웹 페이지 내용을 가져오기
response = requests.get(url)
html_content = response.text

# BeautifulSoup 객체 생성
soup = BeautifulSoup(html_content, 'html.parser')

# 원하는 정보 추출 (제목과 링크)
titles = soup.find_all('h2')  # 원하는 태그명을 넣어주세요
links = soup.find_all('a')     # 원하는 태그명을 넣어주세요

# 추출한 정보 출력
for title in titles:
    print("제목:", title.text.strip())

for link in links:
    print("링크:", link.get('href'))
```
